# Histogram

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
a = np.array([20,87,4,40,53,79,51,11,20,40,15,79,25,27])
plt.hist(a, bins=[0,20,40,60,80,100])
plt.title('Histogram')
plt.show()
```

# Numpy Practical Example

```python
# 1, Plot a Sine Wave
x = np.arange(0, 3*np.pi, 0.1)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
plt.show
```

```python
# Create 6 by 6 array and let 1 and 0s be place alternatively across diagonal.
x = np.zeros((6,6))
x
```

```python
x[1::2,::2] = 1
x
```

```python
x[::2,1::2] = 1
x
```

```python
# Finding Total Number and Locations of Missing values in the arrays

z = np.random.randn(10,10)
z[np.random.randint(10, size=5), np.random.randint(10, size=5)] = np.nan
z
```

```python
# Number of Nan Values
# Note that the out is a boolean
notav = np.isnan(z)

# Printing the total number of the nan Value

print(notav.sum())
```

```python
# Find the exact location of the Nan Values
print(np.argwhere(notav))
```

```python
# Printing the location of the Nans
inds = np.where(notav)
print(inds)
```

```python
# Changing the nan values to Zero
z[inds] = 0
z
```

```python
y =  np.cos(z)
y
```

```python
plt.plot(z, y)
plt.show()
```
